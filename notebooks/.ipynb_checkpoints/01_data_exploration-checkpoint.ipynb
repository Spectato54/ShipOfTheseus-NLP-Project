## Figures
The four figures below describe the corpus for the report's **Dataset** section.
Each figure is also saved to `figures/data_exploration/`.

```python
FIGURES = ROOT / "figures" / "data_exploration"
STAGES = ['T0', 'T1', 'T2', 'T3']
```

## Fig 1: Document Count per Stage per Data Set
Shows how many documents at each stage across the whole corpus.

```python
stage_counts = []

for name, df in datasets.items():
    counts = df.groupby('stage').size().reset_index(name='count')
    counts['dataset'] = name
    stage_counts.append(counts)
stage_df = pd.concat(stage_counts, ignore_index=True)

fig, ax = plt.subplots(figsize=(10, 4.5))
x = np.arange(len(STAGES))
datasets_list = list(datasets.keys())
width = 0.8 / len(datasets_list)
palette = sns.color_palette('muted', len(datasets_list))

for i, (name, color) in enumerate(zip(datasets_list, palette)):
    subset = stage_df[stage_df['dataset'] == name].set_index('stage').reindex(STAGES)
    ax.bar(x + i * width - (len(datasets_list) - 1) * width / 2,
           subset['count'], width=width * 0.9, label=name, color=color)

ax.set_xticks(x)
ax.set_xticklabels(STAGES)
ax.set_xlabel('Paraphrase Stage')
ax.set_ylabel('Document Count')
ax.set_title('Document Count per Stage per Dataset', fontweight='bold')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f'{int(v):,}'))
ax.legend(title='Dataset', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.savefig(FIGURES / 'fig1_stage_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: fig1_stage_distribution.png')
```

## Fig 2: Word Count Distribution
Shows the changing of sentence structure by word count.

```python
datasets_list = list(preprocessed.keys())
ncols = 3
nrows = math.ceil(len(datasets_list) / ncols)

fig, axes = plt.subplots(nrows, ncols,
                         figsize=(6 * ncols, 4.5 * nrows),
                         sharey=False)
axes = axes.flatten()

palette = {"T0" : "#08415C", "T1" : "#CC2936", "T2" : "#CCDBDC", "T3" : "#80CED7"}

for ax, name in zip(axes, datasets_list):
    df_stats = text_stats(preprocessed[name]).copy()
    df_stats['stage'] = preprocessed[name]['stage'].values

    sns.boxplot(
        data=df_stats,
        x='stage', y='word_count',
        hue='stage',
        order=STAGES,
        palette=palette,
        showfliers=False,
        ax=ax
    )

    # Overlay mean line
    means = df_stats.groupby('stage')['word_count'].mean().reindex(STAGES)
    ax.plot(range(len(STAGES)), means.values, 'D--', color='#E53935',
            linewidth=1.8, markersize=7, label='Mean', zorder=5)

    ax.set_title(name, fontweight='bold')
    ax.set_xlabel('Paraphrase Stage')
    ax.set_ylabel('Word Count')
    ax.legend(fontsize=9)

# Hide any unused subplot panels
for ax in axes[len(datasets_list):]:
    ax.set_visible(False)

fig.suptitle('Word Count Distribution by Stage — All Datasets',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(FIGURES / 'fig2_word_count_by_stage.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: fig2_word_count_by_stage.png')
```

## Fig 3: Vocabulary Reduction Across T0 → T3
Compares the raw word count vs. content words across iterations. Wider gap means paraphrasers reduce content words.
Thinner gap means paraphrasers introduce content words.

```python
vocab_rows = []
for name, df in preprocessed.items():
    stats = text_stats(df)
    stats['stage'] = df['stage'].values
    for stage in STAGES:
        subset = stats[stats['stage'] == stage]
        vocab_rows.append({
            'dataset': name,
            'stage': stage,
            'mean_word_count':      subset['word_count'].mean(),
            'mean_no_stop_count':   subset['no_stop_word_count'].mean(),
        })
vocab_df = pd.DataFrame(vocab_rows)

# One panel per dataset
datasets_list = list(preprocessed.keys())
ncols = 3
nrows = math.ceil(len(datasets_list) / ncols)

fig, axes = plt.subplots(nrows, ncols,
                         figsize=(6 * ncols, 4.5 * nrows),
                         sharey=False)
axes = axes.flatten()
if len(datasets_list) == 1:
    axes = [axes]

for ax, name in zip(axes, datasets_list):
    sub = vocab_df[vocab_df['dataset'] == name]
    ax.plot(STAGES, sub['mean_word_count'],    marker='o', linewidth=2.2,
            color='#1565C0', label='All words')
    ax.plot(STAGES, sub['mean_no_stop_count'], marker='s', linewidth=2.2,
            color='#E53935', linestyle='--', label='Content words\n(no stopwords)')
    ax.fill_between(STAGES,
                    sub['mean_word_count'].values,
                    sub['mean_no_stop_count'].values,
                    alpha=0.10, color='#1565C0', label='Stopword gap')
    ax.set_title(name, fontweight='bold')
    ax.set_xlabel('Stage')
    ax.set_ylabel('Mean Word Count')
    ax.legend(fontsize=8.5)
    
# Hide unused panels
for ax in axes[len(datasets_list):]:
    ax.set_visible(False)
    
fig.suptitle('Vocabulary Reduction Across T0 → T3', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(FIGURES / 'fig3_vocabulary_reduction.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: fig3_vocabulary_reduction.png')
```
