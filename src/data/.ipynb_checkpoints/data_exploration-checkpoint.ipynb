## Load the Ship of Theseus Corpus (sci_gen, wp, and xsum)

```python
path = os.path.normpath(os.path.join(globals()["__session__"],"..", "..", "..", "data", "raw"))
sci_gen = "sci_gen_paraphrased.csv"
wp = "wp_paraphrased.csv"
xsum = "xsum_paraphrased.csv"

sci_gen_path = os.path.join(path, sci_gen)
wp_path = os.path.join(path, wp)
xsum_path = os.path.join(path, xsum)

sci_data = pd.read_csv(sci_gen_path)
wp_data = pd.read_csv(wp_path)
xsum_data = pd.read_csv(xsum_path)

datasets = {"sci_gen": sci_data, "wp": wp_data, "xsum": xsum_data}
print("Data Loaded!")
for name, df in datasets.items():
    print("=" * 80)
    print(f"Dataset Shape of {name}: {df.shape}\n")
    print(f"Data Describe: {df.describe()}\n")
    print(f"Column Names: {df.columns.tolist()}")
    print("=" * 80 + "\n")
```

## Build T0 → T3 Chain

```python
source = set(sci_data['source'])
paraphrasers = set(sci_data['version_name'])
paraphraser_orig = [p for p in paraphrasers if p.count('_') == 0 and p != 'original']
chain = ['T0', 'T1', 'T2', 'T3']
def ver_parse(ver):
    if ver == 'original':
        return ver
    v = ver.split('_')
    return v[0]

def ver_map (ver : str):
    if ver == 'original':
        return 'T0'

    # Count underscores
    under_count = ver.count('_')

    if under_count == 0:
        return 'T1'
    elif under_count == 1:
        return 'T2'
    else:
        return 'T3'
        
for data, df in datasets.items():
    df.loc[:,'stage'] = df.loc[:, 'version_name'].apply(ver_map)
    print(f"Stage Distribution ({data}) by Source")
    print(df.groupby('source')['stage'].value_counts())
    print("\n")
    
```
